```python
import  time
import requests
from urllib.request import quote
import pandas as pd
import numpy as np
import json

listing_simple = pd.read_csv('./static/data/listings.csv')
home_lng_lat = listing_simple[['latitude','longitude','id']]
home_lng_lat['longitude'] = home_lng_lat['longitude'].astype("float64")
home_lng_lat['latitude'] = home_lng_lat['latitude'].astype("float64")
```

```python
def GetAddress(pointData):
    url = "http://api.map.baidu.com/reverse_geocoding/v3/"
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.79 Safari/537.36'}
    payload = {
        'output':'json',
        'ak':'bz0r4Ll43eF7Ruhb8Q4Xm7uHQmQ8GpBG'
        }
    addinfo = []
    #计时开始：
    t0 = time.time()
#     print(pointData)
    for index, row in pointData.iterrows():
    # for lon,lat in pointData:
        payload['location'] = '{0:s},{1:s}'.format(str(row[0]),str(row[1]))
        try:
            content =  requests.get(url,params=payload,headers=header).json()
            rs = content['result']
            rs['id'] = row[2]
#             print("rs",rs)
            addinfo.append(rs)
        except:
            addinfo.append(None)
        if(index%100==0):
            t1 = time.time()
            total = t1 - t0
            print("index:{},消耗时间：{}".format(index,total))
#     print("addinfo",addinfo)
    return(addinfo)
```

```python
#保存file，如果有不能转换的格式可以使用NpEncoder函数
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)

```
